# Login in your instagram

```python
# connect webdriver to notebook
driver=webdriver.Chrome('D:/AI/chromedriver.exe')
#specifed which web page do want to open
driver.get('https://www.instagram.com/')
# note : must your chrom in Version 98.0 to run without error
```

```python
'''
when we loading the webpage ,not all elements are loading in the same time,so when we are tring to acess the element it still 
process ,it is not loaded yet ,so the error will appear,selenium can detect this with WebDriverWait,it will wait until 
expected_conditions(EC) the input element e.g username or passward will able to use so we can enter our own info
'''
userName=WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.CSS_SELECTOR,"input[name='username']")))
password=WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.CSS_SELECTOR,"input[name='password']")))
#to make sure the input field is empty
userName.clear()
password.clear()
#to input username
userName.send_keys("#######")
#to input passward
password.send_keys("##########")
```

```python
#we want to select the element of log in & click on it
log_In=WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.CSS_SELECTOR,"button[type='submit']"))).click()
```

```python
#we want to select the buttom not now but buttom is very general so we use xpath insted of CSS_SELECTOR
not_Not=WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH,"//button[contains(text(),'Not Now')]"))).click()
not_Not2=WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH,"//button[contains(text(),'Not Now')]"))).click()
```

```python
#To search on specific something
searchBox=WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH,"//input[@placeholder='Search']")))
#to make sure the input field is empty
searchBox.clear()
#to input username
keyWord="#cat"
searchBox.send_keys(keyWord)
```

```python
#as press enter in keyboard
searchBox.send_keys(Keys.ENTER)
searchBox.send_keys(Keys.ENTER)
```

```python
set_images= set()
```

```python
#scroll down  along the page and select all the images up untill the scrolling point between()we select y location over the scroll
driver.execute_script("window.scrollTo(9000,12000);")
#to select the element from webpage that has tag called img 
images=driver.find_elements_by_tag_name('img')
#now we want to target the source attribute of images
for image in images:
    set_images.add(image
print(len(set_images))
images=[image.get_attribute('src') for image in images]
images
```

```python
for image in set_images:
    print(image)
    break
```

# Save image in computer

```python
#get cuurent directory
path=os.getcwd()
#add a new folder to put the image
path=os.path.join(path,keyWord[1:]+"s")
os.mkdir(path)
path
```

```python
# download the images
counter=0
for image in set_images:
    image=image.get_attribute('src')
    save_As=os.path.join(path,keyWord[1:]+str(counter)+'.jpg')
    wget.download(image,save_As)
    counter+=1
    
```
